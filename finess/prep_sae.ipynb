```python
filtre = (
    pd
    .read_csv(
        'datasets/SAE 2021 Bases statistiques - formats SAS-CSV/Bases statistiques/Bases CSV/FILTRE_2021r.csv',
        sep=';',
        encoding='iso-8859-1',
        dtype='str'
    )
)

filtre
```

```python
heb = [
'HEB_MED',
'HEB_CHIR',
'HEB_PERINAT',
'HEB_PSY',
'HEB_SSR',
'HEB_SLD'
]

no_heb = [
'MED',
'CHIRAMBU',
'PSY',
'RTH',
'CHIMIO',
'IRC',
'IVGAMP',
'CPP',
'HAD',
'SSR',
'URG',
'SMURSAMU'
]

(
    filtre
    .set_index('FI_EJ')
    [heb+no_heb]
    .fillna(0)
    .astype(int)
    .join(finess_ej.set_index('nofiness'))
    .groupby('libstatutjuridique')
    .sum()
)
```
