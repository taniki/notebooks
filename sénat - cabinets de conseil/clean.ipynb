```python
import pandas as pd
import os
```

```python
try:
    os.mkdir('data')
except:
    print('le répertoire `data` existe déja')
```

```python
columns = [
    "numéro",
    "date de début",
    "date de fin",
    "prestataire",
    "période",
    "objet",
    "montant"
]
```

## MAE

```python
mae = (pd.read_excel("files/Affaires_etrangeres_SG.XLSX", header=0, dtype='str')
    .rename(columns={
        "Contrat ou prestation pro bono (libellé)": columns[0],
        "Année de passation": columns[1],
        "Prestataire\n(ex : KPMG, Capgemini, Ernst and Young, etc.)": columns[3],
        "Période couverte\n(ex : 2018-2020)": columns[4],
        'Objet de la prestation\n(ex : "établir un diagnostic des leviers de croissance dans le secteur de…")': columns[5],
        "Montant du contrat\nen euros TTC (AE)": columns[6]
    })
)

mae.dtypes
```

```python
mae.to_csv('data/mae.csv')
```

## Ministère des solidarités et de la santé

```python
mss = (pd.read_excel("files/Affaires_sociales-_SG.xlsx", header=1, dtype='str')
    .rename(columns={
        "Contrat ou prestation pro bono (libellé)": columns[0],
        "Année de passation": columns[1],
        "Prestataire\n(ex : KPMG, Capgemini, Ernst and Young, etc.)": columns[3],
        "Période couverte\n(ex : 2018-2020)": columns[4],
        'Objet de la prestation\n(ex : "établir un diagnostic des leviers de croissance dans le secteur de…")': columns[5],
        "Montant du contrat\nen euros TTC (AE)": columns[6]
    })
)

mss
```

## DINUM

```python
dinum = (pd.read_excel("files/SGG-DINUM.xlsx", header=1, dtype='str')
    .rename(columns={
        "Année de passation : \nDate de notification": "date de notification",
        "Période couverte : \nDate de fin": "date de fin",
        "Prestataire ": "prestataire",
        "Objet de la prestation": "objet",
        "Total général": "montant"
    })
    .astype({
        'prestataire': 'category',
        'objet': 'category',
        'montant': 'float64'
    })
)

dinum.dtypes
```

```python
dinum.to_csv('data/dinum.csv')
```
